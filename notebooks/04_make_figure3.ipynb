# 04 — Generate Figure 3 panels (A–C)

This notebook reproduces Figure 3 panels A–C **with the final, publishable formatting unchanged**.

It runs `scripts/04_make_figure3.py`.

```python
from pathlib import Path
import sys, subprocess, os

# --- locate repo root robustly ---
ROOT = Path.cwd()
if not (ROOT / "scripts").exists() and (ROOT.parent / "scripts").exists():
    ROOT = ROOT.parent
if not (ROOT / "scripts").exists():
    raise RuntimeError("Could not find repo root (missing ./scripts). Start Jupyter from the repo root or the notebooks/ folder.")
print("Repo root:", ROOT)
```

```python
h5ad   = ROOT / "stamp_fig1_scanvi_outputs" / "fig1_scanvi_analyzed.h5ad"
outdir = ROOT / "figures" / "Figure3"
outdir.mkdir(parents=True, exist_ok=True)

cmd = [
    sys.executable, str(ROOT / "scripts" / "04_make_figure3.py"),
    "--h5ad", str(h5ad),
    "--outdir", str(outdir),
]
print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)

print("Saved panels in:", outdir)
```
