# 02 — Build fig1_scanvi_analyzed.h5ad (scVI → scANVI)

This notebook builds the analyzed AnnData object used for Figure 2 and Figure 3.

It runs `scripts/02_scanvi_reference_mapping.py` and writes:
- `stamp_fig1_scanvi_outputs/fig1_scanvi_analyzed.h5ad`

```python
from pathlib import Path
import sys, subprocess, os

# --- locate repo root robustly ---
ROOT = Path.cwd()
if not (ROOT / "scripts").exists() and (ROOT.parent / "scripts").exists():
    ROOT = ROOT.parent
if not (ROOT / "scripts").exists():
    raise RuntimeError("Could not find repo root (missing ./scripts). Start Jupyter from the repo root or the notebooks/ folder.")
print("Repo root:", ROOT)
```

```python
data_dir = ROOT / "data" / "stamp_fig1_samples"
outdir   = ROOT / "stamp_fig1_scanvi_outputs"
outdir.mkdir(parents=True, exist_ok=True)

cmd = [
    sys.executable, str(ROOT / "scripts" / "02_scanvi_reference_mapping.py"),
    "--data_dir", str(data_dir),
    "--outdir", str(outdir),
]
print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)

print("Done. Output:", outdir / "fig1_scanvi_analyzed.h5ad")
```
