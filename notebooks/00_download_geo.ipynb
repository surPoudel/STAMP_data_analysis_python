# 00 â€” Download STAMP GEO supplementary files

This notebook downloads the 4 STAMP Fig1 samples used for the downstream workflow.

It runs the same code as `scripts/00_download_geo.py`.

```python
from pathlib import Path
import sys, subprocess, os

# --- locate repo root robustly ---
ROOT = Path.cwd()
if not (ROOT / "scripts").exists() and (ROOT.parent / "scripts").exists():
    ROOT = ROOT.parent
if not (ROOT / "scripts").exists():
    raise RuntimeError("Could not find repo root (missing ./scripts). Start Jupyter from the repo root or the notebooks/ folder.")
print("Repo root:", ROOT)
```

```python
outdir = ROOT / "data" / "stamp_fig1_samples"
outdir.mkdir(parents=True, exist_ok=True)

cmd = [
    sys.executable,
    str(ROOT / "scripts" / "00_download_geo.py"),
    "--outdir", str(outdir),
    "--max_retries", "6",
    "--timeout", "180",
]
print("Running:", " ".join(cmd))
subprocess.run(cmd, check=True)
```

After this finishes, you should see `.qs.gz`, `_matrix.mtx.gz`, `_barcodes.tsv.gz`, `_features.tsv.gz` in `data/stamp_fig1_samples/`.
